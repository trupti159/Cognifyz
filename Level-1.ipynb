## Task-1 : Top Cuisines

1. Determine the top three most 
common cuisines in the dataset
2. Calculate the percentage of
restaurants that serve each of the to cuisineses.ines.

```python
df.groupby('Cuisines').Cuisines.count()
```

```python
top_cuisines = df['Cuisines'].str.split(',', expand=True).stack().value_counts().head(3)

top_cuisines
```

```python
# plotting the 3 top cuisines

top_cuisines.plot(kind='bar', color=['red','green','orange'])
plt.title('Top 3 Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('No. of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

```python
total_restaurants = len(df)

top_cuisines_percentage = (top_cuisines/total_restaurants)*100

top_cuisines_percentage
```

#### Hence, after analyzing the data 31.35% restaurants serve North Indian cuisines, 19.7% serve chinese and 13.77% serve fast foods.

## Task-2 : City Analysis

1. Identify the city with the highest number 
of restaurants in the dataset
2. Calculate the average rating for 
restaurants in each city
3. Determine the city with the highest 
average rating...

```python
city_restaurants_count = df.groupby('City')['Restaurant ID'].count()
max_restaurant_city = city_restaurants_count.idxmax()
print(f"The city with the maximum restaurants is:",max_restaurant_city)
```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean()
print('the average for restaurants for each city:', city_avg_rating)

sorted_city_avg_rating = city_avg_rating.sort_values(ascending=False)
print('\nCities sorted by average rating:',sorted_city_avg_rating)
```

```python
highest_avg_rating_city = df.groupby('City')['Aggregate rating'].mean().idxmax()
highest_avg_rating_city
```
